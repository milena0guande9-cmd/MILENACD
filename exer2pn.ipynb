```python
import pandas as pd

dados = {
    "Produto": ["Arroz", "Feijao", "Leite", "Pao", "Acucar"],
    "Preco": [50, 60, 40, 10, 45],
    "Quantidade": [20, 15, 30, 50, 25]
}

df = pd.DataFrame(dados)
df

preco_medio = df["Preco"].mean()

print(preco_medio)

produto_mais_caro = df.loc[df["Preco"].idxmax()]

print(produto_mais_caro)

quantidade_total = df["Quantidade"].sum()

print(quantidade_total)
```
