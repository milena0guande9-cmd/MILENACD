```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame 
dados = {
    'Nome': ['Aura', 'Brito', 'Milena', 'Kerry', 'Paco'],
    'Nota': [14.5, 17.0, 19.2, 16.8, 18.0]
}

df = pd.DataFrame(dados)

# 1. Calcular a média das notas
media = df['Nota'].mean()
print(f"Média das notas: {media}")

# 2. Calcular a nota máxima
nota_max = df['Nota'].max()
print(f"Nota máxima: {nota_max}")

# 3. Calcular a nota mínima
nota_min = df['Nota'].min()
print(f"Nota mínima: {nota_min}")

# 4. Usar o método describe
print("\nResumo estatístico das notas:")
print(df['Nota'].describe())

# 5. Criar gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(df['Nome'], df['Nota'], color='skyblue')
plt.title('Notas dos Alunos')
plt.xlabel('')
plt.ylabel('Nota')
plt.ylim(0, 20)  
plt.show()
```
