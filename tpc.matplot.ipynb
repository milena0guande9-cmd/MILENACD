```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "nome": ["Ana", "Bruno", "Carlos", "Dina", "Eduardo"],
    "idade": [23, 35, 29, None, 40],
    "salario": ["50mil", "70mil", "60 mil", "65 mil", None],
    "departamento": ["IT", "HR", "IT", "FN", "IT"]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.describe()
df.info()
```

```python
df["salario"] = (
    df["salario"]
    .str.replace("mil", "", regex=False)
    .str.replace(" ", "", regex=False)
)

df["salario"] = pd.to_numeric(df["salario"], errors="coerce") * 1000
```

```python
df["salario"].dtype
```

```python
maior_salario = df["salario"].max()
df[df["salario"] == maior_salario]
```

```python
df["departamneto"] = df["departamento"].replace("FN", "FIN")
```

```python
plt.figure()
plt.bar(df["nome"], df["salario"])
plt.title("Salário por Funcionario")
plt.xlabel("Nome")
plt.ylabel("Salario")
plt.show()
```

```python
plt.figure()
plt.bar(df["nome"], df["idade"])
plt.title("Idade por Funcionario")
plt.xlabel("Nome")
plt.ylabel("idade")
plt.show()
```
