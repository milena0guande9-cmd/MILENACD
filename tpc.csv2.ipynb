```python
import pandas as pd

data = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana", None],
    "idade": [23, None, 35, 23, 29],
    "cidade": ["Maputo", "Beira", "Maputo", "MAPUTO", "Nampula"],
    "salario": [50000, 60000, None, 50000, 70000]
}

df = pd.DataFrame(data)

df

#
df.to_csv("dados.csv", index=False)
df.head()
df.info()

#valores nulos
df.isnull().sum()

#contagem quantos numeros existem
df.count()

#preenchendo valores nulos
df["nome"] = df["nome"].fillna("Desconhecido")
df["idade"] = df["idade"].fillna(0)
df["cidade"] = df["cidade"].fillna("Sem cidade")
df["salario"] = df["salario"].fillna(0)

#verificacao novamente
df.isnull().sum()
```
